### Question 17:
What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?

```python
mean = np.round(data["Y2017"].mean(), decimals=2)
print(f"mean is {mean}")

median = np.round(data["Y2017"].std(), decimals=2)
print(f"median is {median}")
```

### Question 18:
What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
null = data["Y2014"].isna().sum()
total = len(data)
percentage = (null/total) * 100
print(f"There are {null} missing values")
print(f"The % of missing data is {np.round(percentage, decimals=3)}%")
```
